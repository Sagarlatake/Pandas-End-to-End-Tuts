# Wide to long format

```python
no2_pivited = no2.pivot(index= "date.utc",columns="location", values="value").reset_index()
```

```python
no2_pivited.head()
```

```python
no2_ = no2_pivited.melt(id_vars="date.utc")
```

```python
no2_ = no2_pivited.melt(
        id_vars="date.utc",
        value_vars=["BETR801", "FR04014", "London Westminster"],
        value_name="NO_2",
        var_name="id_location",
)
```
