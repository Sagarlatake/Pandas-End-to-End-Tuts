### Data Cleaning


Data cleaning means fixing bad data in your data set.

Bad data could be:

    Empty cells
    Data in wrong format
    Wrong data
    Duplicates


### Cleaning Empty Cells

```python
df1 = df.copy()
```

```python
df
```

```python
df1.head()
```

```python
df1.isnull()
```

```python
df1.isnull().sum()
```

### Replace Empty Values

Another way of dealing with empty cells is to insert a new value instead.
you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value:

```python
df2 = df.copy()
df2.head()
```

```python
df2.describe()
```

```python
df2.isnull().sum()
```

```python
df2.fillna(25.77, inplace=True)
df2.head()
```

```python
df2.isnull().sum()
```
