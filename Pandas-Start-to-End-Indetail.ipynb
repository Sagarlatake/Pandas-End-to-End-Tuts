### Replace Empty Values

Another way of dealing with empty cells is to insert a new value instead.
you do not have to delete entire rows just because of some empty cells.

The fillna() method allows us to replace empty cells with a value:

```python
df2.describe()
```

```python
df2.isnull().sum()
```

### Mean

```python
data.head()
```

```python
mean1 = data['station_antwerp'].mean()
mean2 = data['station_london'].mean()
mean3 = data['station_paris'].mean()
print(mean1)
print(mean2)
print(mean3)
```

```python
data['station_antwerp'].fillna(mean1, inplace= True)
data['station_london'].fillna(mean2, inplace= True)
data['station_paris'].fillna(mean3, inplace= True)
data.head()
```

### Median

```python
med1 = data['station_antwerp'].median()
med2 = data['station_london'].median()
med3 = data['station_paris'].median()
print(med1)
print(med2)
print(med3)
data['station_antwerp'].fillna(med1, inplace= True)
data['station_london'].fillna(med2, inplace= True)
data['station_paris'].fillna(med3, inplace= True)
data.head()
```

```python
med1 = data['station_antwerp'].mode()[0]
med2 = data['station_london'].mode()[0]
med3 = data['station_paris'].mode()[0]
print(med1)
print(med2)
print(med3)
data['station_antwerp'].fillna(med1, inplace= True)
data['station_london'].fillna(med2, inplace= True)
data['station_paris'].fillna(med3, inplace= True)
data.head()
```

```python
med3 = data['station_paris'].mode()[0]
med3
```
